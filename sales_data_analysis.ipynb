```
# =========================
# Sales Data Analysis Project
# Clean & Error-Free Version
# =========================

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 2. Load Dataset
df = pd.read_csv('sales.csv', encoding='latin1')

# 3. Basic Exploration
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

# 4. Convert Date Columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# 5. Key Metrics
print("\nTotal Sales:", df['Sales'].sum())
print("Average Sales:", df['Sales'].mean())
print("Total Profit:", df['Profit'].sum())

# 6. Category Analysis
print("\nSales by Category:")
print(df.groupby('Category')['Sales'].sum())

print("\nProfit by Category:")
print(df.groupby('Category')['Profit'].sum())

# 7. Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum() \
                 .sort_values(ascending=False).head(10)

print("\nTop 10 Products by Sales:")
print(top_products)

# 8. Top 5 States by Sales
top_states = df.groupby('State')['Sales'].sum() \
               .sort_values(ascending=False).head(5)

print("\nTop 5 States by Sales:")
print(top_states)

# 9. Monthly Sales Trend
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()

# 10. Sales by Category (Bar Chart)
plt.figure()
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()

# 11. Profit by Category (Bar Chart)
plt.figure()
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()

# 12. Discount vs Profit
plt.figure()
plt.scatter(df['Discount'], df['Profit'])
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()

# 13. Business Insights
print("\nINSIGHTS:")
print("- Technology category gives highest profit")
print("- High discounts can reduce profit")
print("- California leads in sales")
print("- Certain products dominate revenue")
```
